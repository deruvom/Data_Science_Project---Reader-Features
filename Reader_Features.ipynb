```python
v_col_names=['visitor_id', 'user_id', 'country', 'timezone', 'location_accuracy']
```

## Question n.1: Sum up readings for horror readers by day

```python
m1 = pd.merge(reading, stories, on='story_id',how='inner')
```

```python
horror_readings=m1[m1['category_one'] =='horror']
```

```python
horror_readings.head()
```

```python
horror_readings=horror_readings[['visitor_id','id','visit_id','tracking_time']]
```

```python
horror_readings['date'] = horror_readings.tracking_time.dt.date
```

```python
horror_readings = horror_readings.groupby('date').agg('count')
horror_readings
```

```python
horror_readings['n_readings/day'] = horror_readings['tracking_time']
```

```python
horror_readings=horror_readings[['n_readings/day']]
```

## Printing the number of readings per day

```python
print(horror_readings)
```
